```python
!ls {path}
```

```python
csv_files_path=os.path.join(path,"ai_job_dataset.csv")
df=pd.read_csv(csv_files_path)
```

```python
df
```

```python
df_da=df[(df['company_location']=='India')& (df['job_title']=='Data Analyst')]

```

```python
df['required_skills']=df['required_skills'].str.split(',').to_list()
```

```python
df_da=df.explode('required_skills')
df_da
```

```python
df_india=df_da.groupby('required_skills')['salary_usd'].agg(['count','median']).sort_values(by='median',ascending=False).head()
df_india
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots(1)
sns.barplot(data=df_india,x=df_india.index,y='median',hue='count',edgecolor='black',palette="Spectral")
ax.set_title('Highest Paid Skills in India')
plt.legend(ncol=10).set_title('Job counts')
ax.invert_yaxis()
plt.tight_layout()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x,_:f'${int(x/1000)}K'))
plt.ylim(0,120000)
plt.xticks(rotation=90)
plt.show()
```
