```python
!pip install IPython
!pip install -qq -U kagglehub

import IPython.display as display
import pandas as pd
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jayaantanaath/student-habits-vs-academic-performance")


```

```python
import os
```

```python
!ls {path}
```

```python
csv_file=os.path.join(path,'student_habits_performance.csv')
df=pd.read_csv(csv_file)
```

```python
df
```

```python
df_social=df[(df['social_media_hours']>4)& (df['exam_score']<40)].sort_values(by='social_media_hours',ascending=False).head()
```

```python
df_social
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots()
sns.barplot(data=df_social,x='social_media_hours',y='exam_score',hue='social_media_hours',palette='light:r')
plt.title('Exam Score vs Social Media Hours')
plt.legend(ncols=3)

plt.show()
```

```python
df_hard_workers=df[(df['part_time_job']=='Yes')& (df['exam_score']>70)&(df['study_hours_per_day']>4)].sort_values(by='exam_score',ascending=False).head()
```

```python
sns.barplot(data=df_hard_workers,x='study_hours_per_day',y='exam_score',hue='part_time_job',palette='blend:b,g')
plt.legend(ncols=1).set_title('Part time Job')
plt.show()
```
