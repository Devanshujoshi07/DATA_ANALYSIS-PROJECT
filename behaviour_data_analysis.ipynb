```python
!pip install IPython
import IPython
import IPython.display as display
!pip install -qq -U kagglehub
import os
import pandas as pd
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jayaantanaath/student-habits-vs-academic-performance")

print("Path to dataset files:", path)
```

```python
!ls {path}
```

```python
csv_file=os.path.join(path,"student_habits_performance.csv")
df=pd.read_csv(csv_file)
df
```

```python
filtered_df = df[(df['study_hours_per_day'] > 5) & (df['exam_score'] > 90)].sort_values(by='exam_score',ascending=False).head()
filtered_df
```

```python
filtered_df_low=df[(df['study_hours_per_day']<2)& (df['exam_score']<40)& (df['gender'])].sort_values(by='exam_score',ascending=False).head()
filtered_df_low
```

```python
import matplotlib.pyplot as plt
```

```python
filtered_df.head().plot(x='study_hours_per_day', y='exam_score', kind='barh')
plt.legend().set_visible(False)
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.title('Scores based on study hours')
plt.show()
```

```python
import seaborn as sns
```

```python
fig,ax=plt.subplots(1,2)

sns.barplot(data=filtered_df,x='study_hours_per_day',y='exam_score',hue='gender',palette='light:g',ax=ax[0])
plt.xlabel('Study Hours per Day')
plt.legend(ncols=2)
plt.ylabel('Exam Score')
plt.tight_layout()
plt.title('Scores based on study hours')

sns.barplot(data=filtered_df_low,x='study_hours_per_day',y='exam_score',hue='gender',palette='light:b',ax=ax[1])
plt.legend(ncols=2)
plt.tight_layout()

plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')



plt.show()

```
